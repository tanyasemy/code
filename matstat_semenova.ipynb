```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
from scipy import stats
import random
```

без поправок

```python
df=pd.DataFrame()
for i in range(20):
    df[i]=np.random.normal(0,4,10000)
```

```python
def ttest(a,b,n):
    mean_diff = np.mean(a) - np.mean(b)
    n1 = len(a)
    n2 = len(b)
    sd_a = np.sqrt(np.var(a,ddof = 1))
    sd_b = np.sqrt(np.var(b,ddof = 1))
    sem = np.sqrt(sd_a**2/n1 + sd_b**2/n2)
    t = mean_diff/sem
    df = (n1 + n2) - 2
    p = stats.t.cdf(-abs(t), df) * 2 
    j=0
    if p > 0.05/n:
        print('Нет статистически значимой разницы','\nt - критерий = ',t,'\np-value =',p)
    else:
        print('Есть статистически значимая разница','\nt - критерий = ',t,'\np-value =',p)
        j=1
    return j
```

```python
n = 10
num = 0
for i in range(n):
    for j  in range(n):
        if i!=j & j < i:
            num = num + ttest(df[i],df[j],n)
print(num)
```

при n<10 - нет пар со значимыми различиями
при n > 10 - есть одна-две пары со значимыми различиями

Поправка Холма - Бонферрони

```python
def ttest1(a,b):
    mean_diff = np.mean(a) - np.mean(b)
    n1 = len(a)
    n2 = len(b)
    sd_a = np.sqrt(np.var(a,ddof = 1))
    sd_b = np.sqrt(np.var(b,ddof = 1))
    sem = np.sqrt(sd_a**2/n1 + sd_b**2/n2)
    t = mean_diff/sem
    df = (n1 + n2) - 2
    return stats.t.cdf(-abs(t), df) * 2 
```

```python
values =[]
n = 20
for i in range(n):
    for j  in range(n):
        if i!=j & j < i:
            values.append(ttest1(df1[i],df1[j]))
```

```python
sort_values = sorted(values)
```

```python
def Holm_Bonferroni(data):
    j=0
    for i in data:
        if i > 0.05/(len(data) - (data.index(i) + 1) + 1):
            print ('Нет статистически значимой разницы','\ncorrected_p_value=',i/(len(data) - (data.index(i) + 1) + 1))
        else:
            print('Есть статистически значимая разница','\ncorrected_p_value=',i/(len(data) - (data.index(i) + 1) + 1))
            j+=1
    print(j)
    return j
```

```python
Holm_Bonferroni(sort_values)/(n*(n-1)/2.0)
```
